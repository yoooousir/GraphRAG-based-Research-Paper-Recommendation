```python
import fitz
def extract_text_from_pdf(pdf_path):
  doc = fitz.open(pdf_path)
  text = ""
  for page in doc:
    text+=page.get_text()
  return text
```

```python
pdf_path = "/content/자연어처리 기반 텍스트 감정 분류 모델.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)
```

```python
from keybert import KeyBERT
def extract_keywords(text, top_n=100):
  kw_model = KeyBERT()
  keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), top_n=top_n)
  return keywords
```

```python
keywords = extract_keywords(extracted_text)
print(keywords)
```

# Graph construction

```python
!pip install networkx transformers
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
def build_graph(keywords, similarity_threshold=0.5):
  G = nx.Graph()
  keyword_phrases = [kw[0] for kw in keywords]
  embeddings = model.encode(keyword_phrases)
  similarity_matrix = cosine_similarity(embeddings)

  #add nodes
  for keyword, _ in keywords:
    G.add_node(keyword)

  #add edges based on co-occurrence or similarity
  num_keywords = len(keywords)
  for i in range(num_keywords):
    for j in range(i+1, num_keywords):
      if similarity_matrix[i, j] >= similarity_threshold:
        # add an edge if similarity is above the threshold
        G.add_edge(keyword_phrases[i], keyword_phrases[j], weight = similarity_matrix[i, j])

  return G
```

```python
def visualize_graph(G):
  plt.figure(figsize=(8,8))
  pos=nx.spring_layout(G)
  nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_color='black', font_weight='bold')
  plt.show()
```

```python
graph = build_graph(keywords, similarity_threshold=0.7)
visualize_graph(graph)
```

# Integration into RAG System

```python
from transformers import pipeline
import networkx as nx
from sentence_transformers import SentenceTransformer

# Step 1: Retrieve relevant text passages (You can use any dense retriever for this, like DPR)
def retrieve_passages(query, text):
    # A simple keyword search can be expanded with embedding-based retrieval
    # Here we'll use keyword matching as a placeholder
    if query in text:
        return text
    return ""

# Step 2: Expand the query using the graph
def expand_query_with_graph(graph, query, top_n=3):
    # Find the most related keywords in the graph
    if query in graph:
        related_keywords = sorted(graph[query], key=lambda x: graph[query][x]['weight'], reverse=True)
        return related_keywords[:top_n]
    return []

# Step 3: Generate an answer using the expanded query and passages
def generate_answer(expanded_query, relevant_passages):
    generator = pipeline("text-generation", model="facebook/bart-large-cnn")
    full_text = relevant_passages + " " + " ".join(expanded_query)
    summary = generator(full_text, max_length=150, do_sample=False)
    return summary[0]['generated_text']

# Example usage
query = "machine learning"
retrieved_passages = retrieve_passages(query, extracted_text)  # Use your document retriever here
expanded_query = expand_query_with_graph(graph, query)

if retrieved_passages:
    answer = generate_answer(expanded_query, retrieved_passages)
    print(answer)
else:
    print("No relevant passages found.")
```
